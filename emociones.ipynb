```python
def encontrar_actividades(columna):
    actividades_unicas = set()

    for fila in columna:
        actividades = fila.split('|')
        actividades = [actividad.strip() for actividad in actividades]
        for actividad in actividades:
            actividades_unicas.add(actividad)
            
    return actividades_unicas


actividades_unicas = encontrar_actividades(df['activities'])

print("Actividades únicas:")
for indice, actividad in enumerate(actividades_unicas, start=1):
    print(f"{indice}. {actividad}")
```
