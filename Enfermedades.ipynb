```python
df_lung_cancer = pd.read_csv('./lung_cancer.csv')
```

```python
df_lung_cancer.head()
```

```python
df_lung_cancer = df_lung_cancer.drop(
    ['YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE', 'WHEEZING', 
     'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN'], axis=1)
```

```python
df_lung_cancer.head()
```

```python
df_lung_cancer['GENDER'] = df_lung_cancer['GENDER'].map({'M': 1, 'F': 0}).astype(int)
df_lung_cancer['SMOKING'] = df_lung_cancer['SMOKING'].map({2: 1, 1: 0}).astype(int)
df_lung_cancer['ALCOHOL CONSUMING'] = df_lung_cancer['ALCOHOL CONSUMING'].map({2: 1, 1: 0}).astype(int)
df_lung_cancer['LUNG_CANCER'] = df_lung_cancer['LUNG_CANCER'].map({'YES': 1, 'NO': 0}).astype(int)
```

```python
df_lung_cancer.info()
```

```python
df_lung_cancer.to_csv('lung_cancer_cleaned.csv', index = False)
```
